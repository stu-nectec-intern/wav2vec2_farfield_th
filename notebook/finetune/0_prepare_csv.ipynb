```python
common_train = pd.read_csv('backup/csv/common_train.csv', index_col=0)
common_test = pd.read_csv('backup/csv/common_test.csv', index_col=0)
common_val = pd.read_csv('backup/csv/common_val.csv' , index_col=0)
```

```python
common_train.head(2)
```

```python
health_df = pd.read_json('backup/Dataset/Healthcare/transcription.json')
smarthome_df =  pd.read_json('backup/Dataset/Smarthome/transcription.json')
```

```python
health_df['path'] = health_df['filename'].apply(lambda x : 'Dataset/Healthcare/Record/' + x + '.wav')
smarthome_df['path'] = smarthome_df['filename'].apply(lambda x : 'Dataset/Smarthome/Record/' + x + ".wav")
```

```python
elder_df = pd.concat([health_df, smarthome_df])
```

```python
elder_df = elder_df.rename(columns= {"transcript":"sentence"})
elder_df = elder_df[["path","sentence"]]
```

```python
elder_train, temp = train_test_split(elder_df, test_size=0.2, random_state=42)
elder_val, elder_test = train_test_split(temp, test_size=0.5, random_state=42)
```

```python
elder_val
```

```python
dataset_train = pd.concat([common_train, elder_train])
dataset_test = pd.concat([common_test, elder_test])
dataset_val = pd.concat([common_val, elder_val])
```

```python
dataset_train
```

```python
def clean_data(data):
    pattern = r"[^ก-๙a-zA-Z0-9\s]+"
    cleaned_data = re.sub(pattern, '', data)
    return cleaned_data
```

```python
dataset_train['sentence'] = dataset_train['sentence'].apply(clean_data)
dataset_test['sentence'] = dataset_test['sentence'].apply(clean_data)
dataset_val['sentence'] = dataset_val['sentence'].apply(clean_data)
```

```python
dataset_train.reset_index(drop=True, inplace=True)
dataset_test.reset_index(drop=True, inplace=True)
dataset_val.reset_index(drop=True, inplace=True)
```

```python
dataset_train['path'] = dataset_train['path'].apply(lambda x : 'backup/' + x)
dataset_test['path'] = dataset_test['path'].apply(lambda x : 'backup/' + x)
dataset_val['path'] = dataset_val['path'].apply(lambda x : 'backup/' + x)
```

```python
import os
import shutil
from concurrent.futures import ThreadPoolExecutor
import pandas as pd
from tqdm import tqdm

def copy_file(row_tuple):
    index, row = row_tuple
    src = row['path']
    dst = row['destination']
    shutil.copyfile(src, dst)
```

```python
dataframes = [dataset_train, dataset_test, dataset_val]

for df in dataframes:
    with ThreadPoolExecutor() as executor:
        list(tqdm(executor.map(copy_file, df.iterrows()), total=df.shape[0]))
```
